```python
import matplotlib.pyplot as plt
import numpy as np

# Data Weighted Scoring Models
criteria = ["Dampak Pendidikan", "Efisiensi Infrastruktur", "Dampak Strategis", 
            "Fleksibilitas", "Biaya dan Sumber Daya"]
weights = [30, 25, 20, 15, 10]  # Bobot dalam persen
scores = [9, 8, 9, 7, 6]  # Nilai per kriteria (1-10)
weighted_scores = [w * s / 10 for w, s in zip(weights, scores)]  # Skor berbobot

# Plot grafik
plt.figure(figsize=(10, 6))
bars = plt.barh(criteria, weighted_scores, color=['blue', 'green', 'red', 'purple', 'orange'])

# Menambahkan nilai pada setiap batang
for bar, score in zip(bars, weighted_scores):
    plt.text(bar.get_width() + 0.2, bar.get_y() + bar.get_height()/2, 
             f"{score:.2f}", va='center', fontsize=12)

# Pengaturan grafik
plt.xlabel("Skor Berbobot")
plt.ylabel("Kriteria")
plt.title("Weighted Scoring Models - Pembangunan Infrastruktur Private Cloud")
plt.xlim(0, max(weighted_scores) + 2)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Menampilkan grafik
plt.show()
```
